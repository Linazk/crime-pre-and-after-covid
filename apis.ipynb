```python
import requests
from pprint import pprint
```

```python
def news_retriever(from_date, to_date, keyword):
    '''
    inputs are start and end date
    keyword is specific keywords to identify only articles with that keyword e.g covid
    orderBy: relevance is the default
    '''

    news_api = f'https://content.guardianapis.com/search?from-date={from_date}&to-date={to_date}&q={keyword}&api-key=ed0c3fd4-7780-417e-aa4f-ff25ddef1c13'
    response = requests.get(url=news_api)
    news_json = response.json()

    return pprint(news_json)


news_retriever('2019-1-10', '2019-2-19', 'breaking-news')
```

```python
# for stringency actions: https://covidtrackerapi.bsg.ox.ac.uk/api/v2/stringency/actions/GBR/2021-09-01

def get_lockdown_state(date):
    '''
    outputs information on the severity of the lockdown or the measures in place by the goverment
    schools closing, restrictions on gatherings etc 
    '''
    covid_api=f'https://covidtrackerapi.bsg.ox.ac.uk/api/v2/stringency/actions/GBR/{date}'
    response_covid = requests.get(url=covid_api)
    covid_json = response_covid.json()
    return pprint(covid_json)

get_lockdown_state(date='2020-09-01') # example date 
```
