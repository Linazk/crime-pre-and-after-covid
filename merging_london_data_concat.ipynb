```python
# sort London 2018, 2019 and 2020 street crime data by 'Month' and 'Reported by'

import pandas as pd

london2018_2020 = pd.read_csv(r"C:\Users\User\d2018_to_2020.csv")

# display original unsorted data
print("\nBefore sorting:")
print(london2018_2020)

# sort data
london2018_2020.sort_values(["Month", "Reported by"],
                            axis=0,
                            ascending=[True, True],
                            inplace=True)

# display sorted data
print("\nAfter sorting:")
print(london2018_2020)
```

```python
# save sorted data

london2018_2020.to_csv('london2018_2020.csv', index=False)
```
